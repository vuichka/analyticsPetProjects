Будем пользоваться двумя методами для приоретизации гипотез: 

$$ 
ICE_{score} = \frac{Impact * Confidence}{Efforts}\\ 
RICE_{score} = \frac{(Reach * Impact * Confidence)}{Efforts}
$$

Заодно проверим, как меняются приоритеты у жтих двух методов.

```python
hypo['ICE'] = (hypo['Impact'] * hypo['Confidence']) / hypo['Efforts'] 
hypo['RICE'] = (hypo['Reach'] * hypo['Impact'] * hypo['Confidence']) / hypo['Efforts'] 
```

Приоритезация задач у двух разных методов довольно существенно различается. Повлиял на это параметр охвата Reach. Метод ICE его не учитывал и поставил на первое место 8 гипотезу, у которой охват равен 1, а метод RICE её учел и на первое место вышла 7 гипотеза с охватом в 10 единиц.

```python
ordersGrouped = orders[['date','group']].drop_duplicates()

ordersCum = ordersGrouped.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'],\
                                                      orders['group'] == x['group'])]\
                .agg({'date': 'max', 'group': 'max', 'revenue': 'sum', 'transactionId': 'nunique'}), axis=1)\
                .sort_values(['date', 'group'])
```

```python
visitorsCum = ordersGrouped.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'],\
                                                                    visitors['group'] == x['group'])]\
                           .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1)\
                           .sort_values(by=['date','group'])
```

```python
cumData = ordersCum.merge(visitorsCum, left_on=['date', 'group'],\
                                 right_on=['date', 'group'])
cumData.columns = ['date', 'group', 'revenue', 'transactions', 'visitors']
cumData
```

```python
fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(111)
sns.lineplot(data=cumData[cumData['group'] == 'A'], x='date', y='revenue', lw=4, color='#808080', ax=ax)
sns.lineplot(data=cumData[cumData['group'] == 'B'], x='date', y='revenue', lw=4, color='#000088', ax=ax)

plt.suptitle('Кумулятивная выручка подвержена выбросам', fontsize=24,\
             x=0.42 ,va='top', color='#000088', y=1)
plt.xlabel('')
plt.ylabel('cumulative revenue', color='#808080')
plt.xticks(rotation=15)

sns.despine()
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(datetime.date(2019,8,31), 4e6, 'Группа A', ha='right',\
        va='top', fontsize=12, color='#808080', rotation=20, clip_on=True)
ax.text(datetime.date(2019,8,31), 5.7e6, 'Группа B', ha='right',\
        va='top', fontsize=12,color='#000088', rotation=25, clip_on=True)

plt.axvline(x=datetime.date(2019,8,18), ymin=0, ymax=0.85, lw=2, color='#616161', alpha=0.3)
plt.annotate('Большой скачок дохода \nу группы Б', xy=(datetime.date(2019,8,18), 6),
             xytext=(datetime.date(2019,8,18), 5400000),
             color='#616161', fontsize=13, clip_on=True)
,
```

           box-shadow: 0 5px 5px rgba(65,96,234,.2);">Статистическая значимость различий в конверсии между группами по «сырым» данным</h2>

Примем уровень статистической значикомсти в 5 процентов, т.е. в 0.05

Проверим гипотезу о том, что конверсии двух групп равны.

H0 - конверсии групп А и В не имеют значимых различий

H1 - конверсии имеют значимые различия

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)


visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders']

sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

print("{0:.3f}".format(ss.mannwhitneyu(sampleA, sampleB)[1]))

print("{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))

print("Значимость критерия Манна-Уитни: {0:.3f}".format(ss.mannwhitneyu(sampleA, sampleB)[1]))

print("Относительный прирост:          {0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

Уровень статистической значимости 5% > 1.1%, следовательно, отвергаем нулевую гипотезу о том, что конверсии двух групп имеют значимые различия.

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > 4]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > 4]['userId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > 60000]['visitorId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormalUsers.head(5))
print(abnormalUsers.shape[0])
```

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
print('Mann-Whitney:  {0:.3f}'.format(ss.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('absolute diff: {0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```
